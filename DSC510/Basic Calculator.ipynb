// Desc: This program is a limited use basic calculator that either 1) performs a calculation using 2 numbers or 2) calculates an average of 2 or more numbers.
// Usage: The program prompts the user to 1) Perform addition, subtraction, multiplication, or division using 2 numbers; 2) Calculate an average of 2+ numbers; or 3) Quit, which ends the program. Both calculator functions are called by the main program via a function. Each function uses a loop, prompts the user for the input they wish to use in the calculation, performs the calculation so long as the input is valid, and provides error messages for invalid input, followed by renewed prompts to obtain the desired input. After each calculation is performed, the user has the option to perform either type of calculation or quit the program.

```python
#Welcome Message
print("WELCOME TO THE PYTHON BASIC CALCULATOR!")

#Global Variables - None

#Function Definitions
#Displays final message and ends the program
def Quit():
    print("\n\nTHANK YOU FOR USING THE PYTHON BASIC CALCULATOR. GOOD BYE!")
        
#Peforms addition, subtraction, multiplication, and division based on user input
def performCalculation(math_symbol):
    while True:
        num1 = input("Please enter the first number of your calculation: ")
        try:
            num1 = float(num1)
        except:
            print("\nERROR: letters, punctuation other than decimals, and other special characters will not be accepted.")
            continue
            
        num2 = input("Please enter the second number of your calculation: ")
        try:
            num2 = float(num2)
        except:
            print("\nERROR: letters, punctuation other than decimals, and other special characters will not be accepted.")
            continue
        
        if math_symbol == '+':
            print("Result:", num1, math_symbol, num2, "=", num1+num2)
            break
        elif math_symbol == '-':
            print("Result:", num1, math_symbol, num2, "=", num1-num2)
            break
        elif math_symbol == '*':
            print("Result:", num1, math_symbol, num2, "=", num1*num2)
            break
        elif math_symbol == '/':
            print("Result:", num1, math_symbol, num2, "=", num1/num2)
            break
        else:
            print("\nERROR: Only +, -, * or / will perform a calculation.")
            break
        
#Gives total sum and average of the numbers entered by the user
def calculateAverage():
    while True:
        SetValues = []
        SetSize = input("How many numbers would you like to enter? ")
        
        try:
            SetSize = int(SetSize)
        except:
            print("\nERROR: Please enter only integers.")
            continue
            
        try:
            for i in range(SetSize):
                SetValues.append(float(input("Please enter a number: ")))
        except:
            print("\nERROR: letters, punctuation other than decimals, and other special characters will not be accepted.")
            continue
            
        total = sum(SetValues)  
        average = sum(SetValues) / SetSize
        print("Total: ", total)
        print("Average: ", average)
        break
            
#Main Program
while True:
    CrossRoad = input("\n\nPlease choose from the following options:\n\n1. +, -, *, or / using 2 numbers\n2. Calculate an average of 2+ numbers\n3. Quit\n\nPlease enter choice 1, 2, or 3: ")

    if CrossRoad == '1':
        math_symbol = input("\n\nWhich calculation would you like to perform? Please enter + - * or / : ")
        performCalculation(math_symbol)
        Calc_Again = input("\nWould you like to perform a new calculation? Please enter Y or N: ")
        if Calc_Again.startswith('n') or Calc_Again.startswith('N'):
            Quit()
            break
        elif Calc_Again.startswith('y') or Calc_Again.startswith('Y'):
            continue
      
    elif CrossRoad == '2':
        calculateAverage()
        Calc_Again = input("\nWould you like to perform a new calculation? Please enter Y or N: ")
        if Calc_Again.startswith('n') or Calc_Again.startswith('N'):
            Quit()
            break
        elif Calc_Again.startswith('y') or Calc_Again.startswith('Y'):
            continue
        break
      
    elif CrossRoad == '3':
        Quit()
        break
```
