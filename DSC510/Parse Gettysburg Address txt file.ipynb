// Desc: This program is a series of 4 functions that 1) opens and reads a .txt file, 2) parses the file string into individual words without punctuation, 3) adds the words to a dictionary with a frequency of 1, and 4) prints both the total word count and each word by frequency.
// Usage: This is a straight-forward program that, as written, is written for only one file: gettysburg.txt. The program prompts the user to enter gettysburg.txt, then automatically runs through the 4 functions described above to print the total word count of the file and also the frequency of each word in the file. The program could be easily adapted to allow other .txt files to be processed in the same way.

```python
#Python Imports
import string

#Global Variable - Creates the Empty Dictionary to Fill w/ Words and Count
gettysburg_dict = {}

#Functions - 4 Functions Including the Main Program

#This function adds words to the dictionary with a frequency of 1. This is called in the next function.
def add_word(word, gettysburg_dict):
    if word in gettysburg_dict:
        gettysburg_dict[word] += 1
    else:
        gettysburg_dict[word] = 1
            
#This function processes each line in the file and strips it of unnecessary characters then calls the add_word function.
def process_line(line, gettysburg_dict):
    line = line.rstrip()
    line = line.translate(line.maketrans('', '', string.punctuation))
    line = line.lower()
    words = line.split()
    for word in words:
        add_word(word, gettysburg_dict)
    
#This function puts the dictionary keys and values into a tuple list, sorts them in reverse order, and prints the output.
def pretty_print(gettysburg_dict):  
    Getty_list=[]
    
    for key, val in gettysburg_dict.items():
        Getty_list.append((val,key))
    
    Getty_list.sort(reverse=True)
    
    print('\nTotal Word Count:\t', len(gettysburg_dict))
    print('\nCount \t Word\n')

    for key, val in Getty_list:
        print(key, '\t', val)

#This function is the main program. It grabs the file as an input, then runs the file through the other functions.        
def main():
    Four_Score = input('Which file would you like to open? ')
    
    try:
        Four_Score = open('gettysburg.txt', 'r')
    except:
        print('ERROR: cannot open the file as entered. The file name you entered is: ', Four_Score)
        exit()
        
    for line in Four_Score:
        process_line(line, gettysburg_dict)
    
    pretty_print(gettysburg_dict)


#Run Main Program
main()
    
    
```
